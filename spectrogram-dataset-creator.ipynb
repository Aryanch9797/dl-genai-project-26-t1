```python
import os
import shutil


KERNEL_SLUG = "aryanchauhan97971234/data-generator-ipynb"
VERSIONS_TO_MERGE = [2, 3, 4, 5, 6]
FINAL_DIR = "final_merged_data"


os.makedirs(FINAL_DIR, exist_ok=True)

print(f"Starting merge")
for version in VERSIONS_TO_MERGE:
    print(f"version:{version}")
    
    temp_path = f"temp_v{version}"
    exit_code = os.system(f"kaggle kernels output {KERNEL_SLUG} -v {version} -p {temp_path}")
    
    if exit_code != 0:
        print(f"Error: Version {version} failed to download. Skipping.")
        continue

    source_train_data = os.path.join(temp_path, "train_data")
    
    if not os.path.exists(source_train_data):
        if os.path.exists(os.path.join(temp_path, "blues")):
            source_train_data = temp_path
        else:
            print(f"Warning: Structure mismatch in Version {version}. Skipping.")
            shutil.rmtree(temp_path)
            continue

    genres = [d for d in os.listdir(source_train_data) if os.path.isdir(os.path.join(source_train_data, d))]
    
    for genre in genres:
        src_genre_path = os.path.join(source_train_data, genre)
        dest_genre_path = os.path.join(FINAL_DIR, genre)
        
        os.makedirs(dest_genre_path, exist_ok=True)

        files = os.listdir(src_genre_path)
        for f in files:
            src_file = os.path.join(src_genre_path, f)
            dest_file = os.path.join(dest_genre_path, f)
            
            shutil.move(src_file, dest_file)
            
    print(f"Version {version} files moved successfully.")

    shutil.rmtree(temp_path)
    print(f"Temp files for Version {version} deleted.")

shutil.make_archive("merged_dataset_complete", 'zip', FINAL_DIR)
shutil.rmtree(FINAL_DIR)
print("data merged")
```
